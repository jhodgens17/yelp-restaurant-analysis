```python
area = []
for num, city in enumerate(df['city']):
    if city in mainland:
        area.append('mainland')
    elif df['latitude'][num] > 39.6:
        area.append('island-north')
    else:
        area.append('island-south')
```

```python
df['area'] = area
```

```python
df.groupby('area').count()['id']
```
